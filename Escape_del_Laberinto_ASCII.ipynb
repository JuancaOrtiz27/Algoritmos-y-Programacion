```python
#Escape del Laberinto ASCII
#Autor: Juan Carlos Ortiz
import os

# Mapa estático (15x15) definido como texto multilínea
mapa_texto = """
###############
#S    #     # #
### # # ### # #
#   #   #   # #
# ##### # # # #
# #   # # #   #
# # # # # ### #
#   # # #   # #
### # # ### # #
#   #       # #
# ### ##### # #
#   #     #   #
### ### # #####
#     # #    E#
###############
"""

# Convertir el mapa en una lista de listas
mapa = [list(linea) for linea in mapa_texto.strip().split('\n')]

# Encontrar posiciones iniciales
def encontrar_posiciones(mapa):
    for y in range(len(mapa)):
        for x in range(len(mapa[0])):
            if mapa[y][x] == 'S':
                inicio = (x, y)
            elif mapa[y][x] == 'E':
                salida = (x, y)
    return inicio, salida

# Validar movimiento
def es_movimiento_valido(mapa, nueva_pos):
    x, y = nueva_pos
    # Verificar límites
    if x < 0 or x >= len(mapa[0]) or y < 0 or y >= len(mapa):
        return False
    # Verificar si es pared
    if mapa[y][x] == '#':
        return False
    return True

# Dibujar mapa
def dibujar_mapa(mapa):
    os.system('cls' if os.name == 'nt' else 'clear')  # Limpiar consola
    for fila in mapa:
        print(''.join(fila))

# Juego principal
def main():
    jugador_pos, salida_pos = encontrar_posiciones(mapa)
    mapa[jugador_pos[1]][jugador_pos[0]] = 'P'  # Colocar jugador en posición inicial

    while True:
        dibujar_mapa(mapa)
        print("Usa WASD para moverte, Q para salir.")
        entrada = input("Movimiento: ").lower()

        if entrada == 'q':
            print("Juego terminado.")
            break

        # Calcular nueva posición
        nueva_pos = jugador_pos
        if entrada == 'w':
            nueva_pos = (jugador_pos[0], jugador_pos[1] - 1)  # Arriba
        elif entrada == 's':
            nueva_pos = (jugador_pos[0], jugador_pos[1] + 1)  # Abajo
        elif entrada == 'a':
            nueva_pos = (jugador_pos[0] - 1, jugador_pos[1])  # Izquierda
        elif entrada == 'd':
            nueva_pos = (jugador_pos[0] + 1, jugador_pos[1])  # Derecha

        # Validar y mover
        if es_movimiento_valido(mapa, nueva_pos):
            # Restaurar celda anterior
            mapa[jugador_pos[1]][jugador_pos[0]] = ' ' if jugador_pos != encontrar_posiciones(mapa)[0] else 'S'
            # Actualizar posición del jugador
            jugador_pos = nueva_pos
            mapa[jugador_pos[1]][jugador_pos[0]] = 'P'

            # Verificar si llegó a la salida
            if jugador_pos == salida_pos:
                dibujar_mapa(mapa)
                print("¡Felicidades! Escapaste del laberinto.")
                break

if __name__ == "__main__":
    main()
```
